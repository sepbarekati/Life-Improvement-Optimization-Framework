# Q1: Appropriate Diet
# Diet Optimization

### DATA:

First, the necessary libraries are imported:

```python
import cvxpy as cp 
import numpy as np
```

The number of food items (N) and the number of nutrients (M) are defined.

Nutrient requirements (B), food costs (C), and the matrix of nutrients present in each food item (P) are defined.


```python

N = 10
M = 5

B = [100.8, 52.1, 370.8, 200.5, 140 ]
C = [45.79, 30.86, 50.2 , 99.26, 32.17, 42.9 , 66.89, 35.06, 13.78, 53.78]
P = [
    [ 70,  5.12,  8.02,  1.89, 12.25],
    [0,  100,  7.62,  2.34, 17.48],
    [ 6.53, 0.1,  7.89,  200,  0.56],
    [0, 0.1,  0.896,  400, 0],
    [18.23,  1.  , 10.29,  9.21,  0.92],
    [ 2.45,  2.93,  2.24,  4.55,  8.87],
    [ 9.41, 13.1 ,  7.24,  0.24,  7.35],
    [19.21,  1.32, 12.35, 11.22,  5.14],
    [ 1.27, 11.66,  256,  1.14,  2.42],
    [11.37, 17.86,  33, 0,  100]
    ]
```

Modeling

```python
# Optimization variables representing the quantity of each food
x = cp.Variable(N, nonneg=True)

# Objective function: Minimize the total cost of food
objective = cp.Minimize(C @ x)

# Constraints: Nutritional requirements must be met
constraints = [np.transpose(P) @ x >= B]

# Solving the optimization problem
problem = cp.Problem(objective, constraints)
problem.solve()
```

Results

```python
# Displaying the optimal status, total cost, and consumption of each food
print("Optimal Status:", problem.status)
print("Total Cost:", problem.value)
print("Consumption of each food:", x.value)
```

# Q2: Project Management

### DATA:

```python
N = 5
P = 2500
t = 800
R = 200
C = [2,  2.09,  4.45, 5.55,  1.52]
T = [3.57,  2.448, 2.69, 3.03,  9.55]
p = [100,8,35,90,20]
```

Modeling

```python
# Optimization variables are defined
x = cp.Variable(N, nonneg=True)

objective = cp.Minimize(C @ x)

constraints = [T @ x <= t, cp.sum(x) <= R, p @ x >= P]

problem = cp.Problem(objective, constraints)
problem.solve()
```

```python
print("Optimal status:", problem.status)
print("Total cost:", problem.value)
print("Resource allocation to each activity:", x.value)
```

# Q3: Portfolio Selection
# Investment Portfolio Optimization

### DATA:

Data for expected returns (mu), covariance matrix (Sigma), and the risk-aversion weight coefficient (lambda) are defined.


```python
mu = [19.12, 13.31, 18.4 , 19.56, 18.7 , 14.04, 15.21, 17.74, 19.88, 19.39]

Sigma = [
    [3.44, 2.76, 2.28, 2.35, 2.83, 1.98, 2.22, 2.65, 3.37, 2.49],
    [2.76, 3.68, 2.74, 2.26, 3.06, 1.73, 2.38, 2.16, 2.79, 2.13],
    [2.28, 2.74, 2.62, 1.66, 2.72, 1.43, 2.28, 2.04, 2.35, 1.8 ],
    [2.35, 2.26, 1.66, 2.88, 1.88, 2.14, 2.23, 2.12, 2.59, 2.71],
    [2.83, 3.06, 2.72, 1.88, 3.51, 1.69, 2.34, 2.48, 3.06, 2.13],
    [1.98, 1.73, 1.43, 2.14, 1.69, 2.82, 1.64, 2.57, 2.81, 2.26],
    [2.22, 2.38, 2.28, 2.23, 2.34, 1.64, 2.82, 1.92, 2.33, 2.57],
    [2.65, 2.16, 2.04, 2.12, 2.48, 2.57, 1.92, 3.06, 3.19, 2.24],
    [3.37, 2.79, 2.35, 2.59, 3.06, 2.81, 2.33, 3.19, 3.9 , 2.8 ],
    [2.49, 2.13, 1.8 , 2.71, 2.13, 2.26, 2.57, 2.24, 2.8 , 3.26]
       ]

lambda_ = 0.5
```

```python
# Number of assets
n = len(mu)
# Optimization variables representing the share of each asset in the portfolio
x = cp.Variable(n)

# Objective function to maximize expected return minus the weighted risk
objective = cp.Maximize(mu @ x - lambda_ * cp.quad_form(x, Sigma))

# Constraints: The sum of shares must be equal to 1, and each share must be between 0 and 1
constraints = [cp.sum(x) == 1, x >= 0, x <= 1]

# Define and solve the optimization problem
problem = cp.Problem(objective, constraints)
problem.solve()
```

```python
# Displays the optimality status of the problem
print("Optimal status:", problem.status)
# Displays the expected return of the portfolio
print("Expected return:", (mu @ x).value)
# Displays the risk of the portfolio
print("Risk:", cp.quad_form(x, Sigma).value)
# Displays the investment amount in each asset
print("Investment in each asset:", x.value)
```
